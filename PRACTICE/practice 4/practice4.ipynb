```python
import requests
url = "https://api.polygon.io/v2/aggs/ticker/HBI/range/1/minute/2022-08-01/2022-08-31?adjusted=true&sort=asc&limit=50000&apiKey=BnVysXNFQpDFB6zKnDmp4jyvtpRla8G2"
resp = requests.get(url)
print(resp.json())
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(resp.json()['results'])
```

```python
print(df)
```

```python
import datetime
def change_time(x):
    return datetime.datetime.fromtimestamp(x/1000.0)

df['time'] = df['t'].apply(change_time)
```

```python
print(df)
```
