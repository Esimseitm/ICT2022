```python
#ВАША ОЧЕРЕДЬ (15 минут)
#Найти все пропущенные значения в предоставленном фрейме данных
#Проверить все нули
#Найти другие типы пропущенных значений
#Есть '-', 'ОШИБКА', неверная дата, 'NaN', нет данных
#а потом сломать
import datetime
a = ['ERROR', '-']
df = pd.read_csv(r"C:\Users\Manarbek\Desktop\ICT\LECTURE\Lecture_7\Apple_price_to_clean.csv", na_values = a)
print(df)
for index, lines in df.iterrows():
    try:
        a = datetime.datetime.strptime(lines['Date'], '%Y-%m-%d')
    except:
        print(lines['Date'])
        df.loc[index, 'Date'] = np.nan
print(df)
```

```python
## Replacing missing values замена пропущенных значений
#удалить строки с дропной с аргументами subset и inplace
df.dropna(inplace = True, subset = ['Low'])
# subset удалает только в тех столбцах 
print(df)
```

```python
# Замена на основе местоположения с помощью функции loc
df.loc[2, 'ST_NUM'] = 200
print(df)
```

```python
# Replace missing values with a number (fillna() function with inplace arg)
# Замените отсутствующие значения числом (функция fillna() с аргументом на месте)
df.fillna(100, inplace = True)
print(df)
```

```python
#Замена отсутствующих значений медианой
#(вычислить медиану, а затем использовать замену числом)
#Replacing the missing values with a median 
#(calculate median and then use the replace with a number)
print(df['Volume'].mean())
df['Volume'].fillna(df['Volume'].mean(), inplace = True)
print(df)
```

```python
# Замена значений на одно до или после: df.fillna(method='bfill')
print(df)
df.fillna(method = 'bfill', inplace = True)
print(df)
```

```python
#Use the drop() function with inplace and axis arg
#axis 0 => lines
#axis 1 => columns
#Используйте функцию drop() с inplace и аргументом оси
#ось 0 => линии
#ось 1 => столбцы
print(df)
df.drop(['ST_NUM'], axis = 1, inplace = True)
print(df)
```

```python
#Используйте дубликаты для удаления ненужных строк
df.drop_duplicates(inplace = True)
print(df)
```

```python
import pandas as pd
dataframes = []
i = 0;
cnt = 0
while cnt != 10:
    try:
        url = "https://query1.finance.yahoo.com/v7/finance/download/{}?period1=0&period2=1661904000&interval=1d&events=history&includeAdjustedClose=true".format(l[i])
        df = pd.read_csv(url)
        df.insert(7, 'Company Symbol', l[i])
        dataframes.append(df)
        i+=1
        cnt+=1;
    except:
        i+=1
        continue;
print(dataframes)    
```
